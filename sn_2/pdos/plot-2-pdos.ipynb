```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline

# load data
def data_loader(fname):
    import numpy as np

    data = np.loadtxt(fname)
    energy = data[:, 0]
    pdos = data[:, 1]

    return energy, pdos

energy, pdos_s = data_loader('atom_Sn_orbital_s.dat')
_, pdos_p = data_loader('atom_Sn_orbital_p.dat')
_, pdos_d = data_loader('atom_Sn_orbital_d.dat')
_, pdos_tot = data_loader('stanene.pdos_tot')

# make plots
plt.figure(figsize = (8, 4))
plt.plot(energy, pdos_s, linewidth=0.75, color='#006699', label='s-orbital')
plt.plot(energy, pdos_p, linewidth=0.75, color='r', label='p-orbital')
plt.plot(energy, pdos_d, linewidth=0.75, color='g', label='d-orbital')
plt.plot(energy, pdos_tot, linewidth=0.75, color='k', label='total')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x= -2.1246, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(-5, 5)
plt.ylim(0, 4)
plt.fill_between(energy, 0, pdos_s, where=(energy < -2.1246), facecolor='#006699', alpha=0.25)
plt.fill_between(energy, 0, pdos_p, where=(energy < -2.1246), facecolor='r', alpha=0.25)
plt.fill_between(energy, 0, pdos_tot, where=(energy < -2.1246), facecolor='k', alpha=0.25)
# plt.text(6.5, 0.52, 'Fermi energy', fontsize= small, rotation=90)
plt.legend(frameon=False)
plt.show()
```
