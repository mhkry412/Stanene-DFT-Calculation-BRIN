```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
plt.style.use('../../matplotlib/sci.mplstyle')
import numpy as np

# The Fermi energy, find it in nscf.out
efermi = -2.1246 

# Define the function to read the data file
def r_dos(name):
    ener, dos = np.loadtxt(name, usecols=(0,1), unpack=True)
    return ener, dos

# Open and read the total PDOS
ener, dos = r_dos('stanene.pdos_tot')
# Open and read the PDOS files for Sn atom number 1
ener1, p1Sn1d = r_dos('stanene.pdos_atm#1(Sn)_wfc#1(d)')
ener1, p1Sn2s = r_dos('stanene.pdos_atm#1(Sn)_wfc#2(s)')
ener1, p1Sn3p = r_dos('stanene.pdos_atm#1(Sn)_wfc#3(p)')
# Open and read the PDOS files for Sn atom number 2
ener2, p2Sn1d = r_dos('stanene.pdos_atm#2(Sn)_wfc#1(d)')
ener2, p2Sn2s = r_dos('stanene.pdos_atm#2(Sn)_wfc#2(s)')
ener2, p2Sn3p = r_dos('stanene.pdos_atm#2(Sn)_wfc#3(p)')

# Create figure object
plt.figure()
# Plot the DOS
plt.plot(ener-efermi, dos, c='k')
# Plot the PDOS of d-orbital
plt.plot(ener1-efermi, p1Sn1d+p2Sn1d, c='g')
# Plot the PDOS of s-orbital
plt.plot(ener1-efermi, p1Sn2s+p2Sn2s, c='b')
# Plot the PDOS of p-orbital
plt.plot(ener2-efermi, p1Sn3p+p2Sn3p, c='r')
# Add the x and y-axis labels
plt.xlabel('Energy (eV)')
plt.ylabel('PDOS (state/eV/unit-cell)')
# Set the axis limits
plt.xlim(-10, 2)
plt.ylim(0, 4)
# Save the figure
plt.savefig('plot-pdos.pdf')
# Show the figure
plt.show()
```
